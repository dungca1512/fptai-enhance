```python
import pandas as pd
import requests
import json
from tqdm import tqdm
import numpy as np
import time
tqdm.pandas()
```

```python
def json_read(path: str):
    with open(path, encoding="utf-8") as fh:
        data = json.load(fh)
    return data
```

```python
labels = pd.read_csv("20230630_094632.csv")
filenames = labels["File Name"].unique().tolist()
```

```python
call_data = pd.read_json("m3_call_clean.json")
call_data = call_data[call_data["name"].isin(filenames)]
call_data.sample()
```

```python
wc_project_info = json_read("m3_project_info.json")["project_info"]
# criteria_name = ["hc_clx_m3_oh1", "hc_clx_m3_oh2"]
criteria_name = ["hc_clx_m3_askpaymentdatetime"]
```

```python
def test_criteria(project_info, criteria_name, transcript, metadata, agent_channel, file_name="test"):
    url = "http://103.176.146.250:5005/predict/dialogue/test"
    payload = json.dumps({
      "project_info": project_info,
      "transcript": transcript,
      "metadata": metadata,
      "criteria": criteria_name,
      "fileName": file_name,
      "agentChannel": agent_channel
    })
    headers = {
      'Content-Type': 'application/json'
    }
    response = requests.request("POST", url, headers=headers, data=payload)
    return response.json()

def calculate_criteira(rec, wc_project_info, criteria_name):
    while True:
        try:
            result = test_criteria(wc_project_info, criteria_name,
                           rec["content"], rec["metadata"],
                           rec["agentChannel"], file_name=rec["name"])
            for criteria in criteria_name:
                rec[criteria] = result[criteria]["evaluate"]
            break
        except:
            print(result.keys())
            print(rec["name"])
            time.sleep(10)
    return rec
```

```python
sample = call_data.sample().to_dict("records")[0]
result = test_criteria(wc_project_info, criteria_name, sample["content"], sample["metadata"], sample["agentChannel"], file_name=sample["name"])
```

```python
result
```

```python
call_data = call_data.progress_apply(lambda rec: calculate_criteira(rec, wc_project_info, criteria_name), axis=1)
```

```python
compare = labels.loc[labels["Criterion Name"] == "When"].merge(call_data, left_on='File Name', right_on='name')
compare = compare[["name", "Criterion Name", "Expected Result"] + criteria_name]
compare['Expected Result'] = compare['Expected Result'].replace(np.nan, "no")
compare["Expected Result"] = compare["Expected Result"].apply(str.lower)
```
